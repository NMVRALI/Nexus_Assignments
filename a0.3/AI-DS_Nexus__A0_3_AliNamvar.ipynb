## 1. Log and Exp: Proving Inverse Relationship


```python
import numpy as np

# Show log(exp(x)) == x and exp(log(x)) == x for several values
x_vals = np.array([1, 2, 10, np.e])
print("x\tlog(exp(x))\texp(log(x))")
for x in x_vals:
    print(f"{x:.2f}\t{np.log(np.exp(x)):.2f}\t\t{np.exp(np.log(x)):.2f}")

# Edge case: try negative and zero values (show error/NaN for log)
y_vals = np.array([0, -1, 5])
print("\nlog(exp(y)) for 0 and negative values:")
print(np.log(np.exp(y_vals)))
```

### Observation üëÄ

- `log(exp(x))` = x ‚úÖ  (works for all real numbers)
- `exp(log(x))` = x ‚ö†Ô∏è  (only works if x > 0)


## 2. Plotting Sine Functions Variations



```python
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 500)

# look at the description of each function and call the proper function from numpy
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.sin(2*x), label='sin(2x)')
plt.plot(x, np.sin(x/2), label='sin(x/2)')
plt.plot(x, np.sin(x)**2, label='sin¬≤(x)')
plt.legend()
plt.title("Variations of Sine Functions")
plt.show()
```

### üëÄüîé What do you see?

### Observation üëÄ

- The plot shows **four variations of the sine function**:

1. `sin(x)` ‚Äì standard sine wave.
2. `sin(2x)` ‚Äì **double frequency**, more oscillations.
3. `sin(x/2)` ‚Äì **half frequency**, stretched horizontally.
4. `sin¬≤(x)` ‚Äì always **non-negative**, oscillates between 0 and 1.

- Scaling `x` inside `sin()` changes the **frequency**.
- Squaring `sin(x)` changes the **amplitude pattern**.

## 4. Features of the Sigmoid Function

```python

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10, 200)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("œÉ(x)")
plt.grid(True)
plt.show()

print("Range:", (y.min(), y.max()), "| At x=0:", sigmoid(0))
```

# Sigmoid Function Plot

This is a plot of the **Sigmoid function**, a common activation function in neural networks.

- **x-axis:** represents the input \(x\)  
- **y-axis:** represents \(\sigma(x)\), which is the sigmoid output  

The curve smoothly transitions from 0 to 1, with a **midpoint at \(x = 0\)** where \(\sigma(0) = 0.5\).  

The function has an **S-shaped (sigmoid) form**, approaching:
- 0 as \(x \to -\infty\)  
- 1 as \(x \to +\infty\)


## 5. Derivatives of Famous Functions

```python
x = np.linspace(-3, 3, 400)
plt.plot(x, x**2, label="x¬≤")
plt.plot(x, 2*x, '--', label="d/dx x¬≤")

x = np.linspace(0, 3, 400)
plt.plot(x, np.exp(x), label="exp(x)")
plt.plot(x, np.exp(x), '--', label="d/dx exp(x)")

plt.plot(x, np.log(x+0.1), label="log(x)")
plt.plot(x,  1/(x + 0.1), '--', label="d/dx log(x)")

plt.legend()
plt.title("Functions and Their Derivatives")
plt.show()
```

# Functions and Their Derivatives

This plot compares three functions with their derivatives:

- **x¬≤ (blue)** is a parabola opening upward.  
  - Its derivative **2x (orange dashed)** is a straight line, showing the slope increases linearly.

- **exp(x) (green)** grows rapidly as x increases.  
  - Its derivative **exp(x) (red dashed)** is identical, meaning the function‚Äôs rate of change equals its value.

- **log(x) (purple)** is defined only for x > 0 and increases slowly.  
  - Its derivative **1/x (brown dashed)** decreases as x increases, showing the slope becomes smaller.

Overall, the plot visualizes how each derivative represents the slope behavior of its original function.


## 6. Gradient of Selected Functions

```python
# search and learn more about sumpy
import sympy as sp

# Define symbolic variables
x, y = sp.symbols('x y')

# Define the function
f = x**2 * y + sp.sin(y)

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Function: f(x, y) =", f)
print("Partial derivative w.r.t x (‚àÇf/‚àÇx):", df_dx)
print("Partial derivative w.r.t y (‚àÇf/‚àÇy):", df_dy)
```

# Symbolic Multivariable Calculus with SymPy

We create a function **symbolically** using SymPy:

$$f(x, y) = x^2 y + \sin(y)$$

---

## Partial Derivatives

**Partial derivative with respect to \( x \):**

$$\frac{\partial f}{\partial x} = 2 x y$$

**Partial derivative with respect to \( y \):**

$$\frac{\partial f}{\partial y} = x^2 + \cos(y)$$

---

## Key Points

This example shows how **SymPy** can:

- Represent functions **symbolically**  
- Compute **exact derivatives**  
- Handle **multivariable calculus analytically** (not numerically)
